```python
import pandas as pd
```

```python
df = pd.read_csv('State_LGA_Federal_Locality_Postcode.csv', index_col=False)
```

```python
df.drop_duplicates(subset=['Postcode', 'State District'], inplace=True)
df['Postcode'] = df['Postcode'].astype(str)
```

```python
df
```

```python
district_group = df.groupby('State District')['Postcode'].apply(list)
district_group
```

```python
region_name_postcodes = pd.DataFrame({'name':district_group.index, 'postcodes':district_group.values})
region_name_postcodes
```

```python
region_name_postcodes.to_json('region_name_postcodes.json', orient='records')
```
