```python
import pandas as pd
```

```python
df = pd.read_csv('State_LGA_Federal_Locality_Postcode.csv', index_col=False)
df['Postcode'] = df['Postcode'].astype(str)
```

# State Regions

```python
state_df = df.drop_duplicates(subset=['Postcode', 'State District'])
state_df
```

```python
state_postcodes = state_df.groupby('State District')['Postcode'].apply(list)
```

```python
state_suburbs = state_df.groupby('State District')['Locality'].apply(list)
```

```python
state_regions = pd.DataFrame({'name':state_postcodes.index, 'type': 'State', 'postcodes':state_postcodes.values, 'suburbs':state_suburbs.values})
state_regions
```

# Federal Regions

```python
df.sort_values(by=['Federal District'], inplace=True)
federal_df = df.drop_duplicates(subset=['Postcode', 'Federal District'])
federal_df
```

```python
federal_postcodes = federal_df.groupby('Federal District')['Postcode'].apply(list)
federal_postcodes
```

```python
federal_suburbs = federal_df.groupby('Federal District')['Locality'].apply(list)
federal_suburbs
```

```python
federal_regions = pd.DataFrame({'name':federal_postcodes.index, 'type': 'Federal', 'postcodes':federal_postcodes.values, 'suburbs':federal_suburbs.values})
federal_regions
```

# Append the data

```python
all_regions = pd.DataFrame()
all_regions.append([state_regions, federal_regions])
all_regions
```
